```python
import pandas as pd

# Ensure datetime columns
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date'
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Mask for incorrect chronological order (ignoring NaNs)
incorrect_mask = (
    (df['order_approved_at'].notna() & df['order_purchase_timestamp'].notna() & 
     (df['order_approved_at'] < df['order_purchase_timestamp'])) |
    (df['order_delivered_carrier_date'].notna() & df['order_approved_at'].notna() &
     (df['order_delivered_carrier_date'] < df['order_approved_at'])) |
    (df['order_delivered_customer_date'].notna() & df['order_delivered_carrier_date'].notna() &
     (df['order_delivered_customer_date'] < df['order_delivered_carrier_date']))
)

# Filter incorrect rows
incorrect_rows = df[incorrect_mask]

# Show results
incorrect_rows.head(10)
```

```python
df1 = df.copy()

date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

df1 = df1.apply(lambda col: col.str.strip() if col.dtype == "object" else col)

for col in date_cols:
    df1[col] = pd.to_datetime(df1[col])

df1 = df1[df1['order_status'] == 'delivered'].copy()

df1['order_date'] = df1['order_purchase_timestamp'].dt.date
df1['delivery_date'] = df1['order_delivered_customer_date'].dt.date

df1['delivery_days'] = (df1['order_delivered_customer_date'] - df1['order_purchase_timestamp']).dt.days

df1['is_delayed'] = df1['order_delivered_customer_date'] > df1['order_estimated_delivery_date']
print(df1.isna().sum())
```
